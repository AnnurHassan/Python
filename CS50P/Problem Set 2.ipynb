```python
# camelCase

def main():
    camelCase = input("Please enter a variable name is camelCase : ").strip()
    snake_case = ""

    for i in camelCase:

        if i.isupper():
            snake_case += "_"
            snake_case += i.lower()

        else:
            snake_case += i

    print(snake_case)


main()
```

```python
# Coke Machine

def main():
    coke = 50
    coin = int(input("Insert Coin: "))

    if coin == 25 or coin == 10 or coin == 5:
        due = coke - coin
        print(f'Amount Due: {due}')
    else:
        due = coke
        print(f'Amount Due: {due}')

    while True:

        due_amount = int(input("Insert Coin"))
        due = due - due_amount

        if due == 0:
            print(f'Change Owed: {due}')
            break

        elif due < 0:
            print(f'Change Owed: {abs(due)}')
            break

        else:
            print(f'Amount Due: {due}')
            continue


main()
```

```python
# Just setting up my Twttr

def main():
    string = input("Input: ")
    new_string = ""

    for i in string:

        if isvowel(i):
            continue

        else:
            new_string += i

    print(f"Output: {new_string}")


def isvowel(chr):
    vowels = "aeiouAEIOU"

    if chr in vowels:
        return True
    
    else:
        return False


main()
```

```python
# Vanity Plates

def main():
    plate = input("Plate: ").strip()

    if is_valid(plate):
        print("Valid")

    else:
        print("Invalid")


def is_valid(s):
    rem = ""
    if first_two_check(s) and is_accurate_length(s) and no_puntuation(s):
        if s[-1].isalpha():

            for i in s:
                if i.isnumeric():
                    return False
                else:
                    continue
            return True

        else:
            for i in range(2,len(s)):
                check = s[i]
                if check.isnumeric():
                    if check == "0":
                        if rem == "":
                            return False
                        else:
                            rem = "num"
                            continue
                    else:
                        rem = "num"
                        continue
                else:
                    if rem == "num":
                        return False
                    else:
                        continue
            return True
    else:
        return False



def first_two_check(plate):
    check = plate[0:2]

    if check.isalpha():
        return True
    else:
        return False


def is_accurate_length(plate):

    if len(plate) >= 2 and len(plate) <= 6:
        return True
    else:
        return False


def no_puntuation(plate):
    puntuation = [" ", ", ", ".", ":"]

    for i in plate:
        if i in puntuation:
            valid = False
            break
        else:
            valid = True

    return valid


main()
```
