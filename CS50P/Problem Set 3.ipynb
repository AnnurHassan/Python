```python
# Fuel Gauge

def main():
    x,y = verify_input()
    fuel = (x / y) * 100
    if fuel <= 1:
        print("E")
    elif fuel >= 99:
        print("F")
    else:
        print(f'{round(fuel)}%')


def verify_input():
        while True:
            try:
                fraction = input("Please enter a fraction : ").strip().split("/")
                x,y = fraction
                x = int(x)
                y = int(y)
            except ValueError:
                print("""
                Enter fraction in the correct Format (x/y)
                X and Y has to be integers.
                """)
            else:
                if x <= y and y != 0 :
                    break
                else:
                    continue

        return x,y


main()
```

```python
#Felipe's Taqueria

def main():
    total = 0
    while True:
        try:
            item = input("Please enter name of the item: ").strip().lower()
        except EOFError:
            print()
            break
        except KeyboardInterrupt:
            print()
            break
        else:
            if not get_price(item):
                continue
            else:
                price = get_price(item)
                total += price
                print(f'Total: ${total:.2f}')



def get_price(item):
    database = {
    "baja taco": 4.00,
    "burrito": 7.50,
    "bowl": 8.50,
    "nachos": 11.00,
    "quesadilla": 8.50,
    "super burrito": 8.50,
    "super quesadilla": 9.50,
    "taco": 3.00,
    "tortilla salad": 8.00
    }

    try:
        price = database[item]
        return price
    except KeyError:
        return False


main()
```

```python
# Grocery List

def main():
    item_list = create_list()
    item_list.sort()
    item_dict = {}
    for i in item_list:
        count = item_list.count(i)
        if i in item_dict.keys():
            continue
        else:
            item_dict[i.upper()] = count

    for key,value in item_dict.items():
        print(f'{value} {key}')


def create_list():
    item_list = []
    while True:
        try:
            item = input()
        except EOFError:
            print()
            break
        except KeyboardInterrupt:
            print()
            break
        else:
            item_list.append(item)
    return item_list


main()
```

```python
#Outdated

month_list = [
    "January",
    "February",
    "March",
    "April",
    "May",
    "June",
    "July",
    "August",
    "September",
    "October",
    "November",
    "December"
]

def main():
    month, day, year = get_input()
    if (1 <= month <=9) and (1 <= day <= 9):
        print(f'{year}-0{month}-0{day}')
    elif (1 <= month <=9) and not(1 <= day <= 9):
        print(f'{year}-0{month}-{day}')
    elif not(1 <= month <=9) and (1 <= day <= 9):
        print(f'{year}-{month}-0{day}')
    else:
        print(f'{year}-{month}-{day}')

def get_input():
    while True:
        date = input("Date: ").strip()
        if "," in date:
            month, day, year = date.split(" ")
            try:
                day = int(day.removesuffix(","))
            except:
                continue
            else:
                if (month in month_list) and (1 <= day <= 31):
                    month = (month_list.index(month)) + 1
                    break
                else:
                    continue

        elif "/" in date:
            month, day, year = date.split("/")
            try:
                month = int(month)
                day = int(day)
            except:
                continue
            else:
                if (1 <= month <= 12) and (1 <= day <= 31):
                    break
                else:
                    continue
        else:
            continue

    return month, day, year

main()
```
